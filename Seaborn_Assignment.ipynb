## 2: Plot 12 most dangerous cities

```python
ctyy= kill.city.value_counts().head(12)
ctyy
```

```python
yvalue= ctyy.values
yvalue
```

```python
xindex = ctyy.index
xindex
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x=xindex, y= yvalue)
ax.set_xticklabels(ctyy.index, rotation=45, color="b")
ax.set_title("Most Dangerous Cityes", fontsize=14, color="r")

plt.tight_layout()
```

```python
kill.city.value_counts().head(12)
```

```python
xindex = ctyy.index
xindex
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='city', data=kill, order= xindex)
ax.set_xticklabels(ctyy.index, rotation=45, color="b")
ax.set_title("Most Dangerous Cityes", fontsize=14, color="r")

plt.tight_layout()
```
